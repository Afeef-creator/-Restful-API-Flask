- Question 1 : What is a RESTful API?

          - A RESTful API is a way for applications to communicate over the internet using HTTP.
          - It uses methods like GET, POST, PUT, and DELETE to perform different actions.
          - It is stateless, meaning each request is handled independently.
          - It mainly uses JSON for sending and receiving data.
          - REST APIs are simple, flexible, and widely used in modern web development.



- Question 2 : What is Flask, and why is it popular for building APIs?

          - Flask is a lightweight Python web framework used to build web apps and APIs.
          - It is popular because it is simple to learn and very flexible.
          - Flask requires very little setup, making it fast to develop APIs.
          - It supports many useful extensions like Flask-SQLAlchemy.
          - Developers prefer Flask because it gives full control over the project structure.



- Question 3 : What are HTTP methods used in RESTful APIs?

          - GET is used to fetch or read data.
          - POST is used to create new data on the server.
          - PUT is used to fully update existing data.
          - PATCH is used to partially update a resource.
          - DELETE is used to remove data from the server.



- Question 4 : What is the purpose of the @app.route() decorator in Flask?

          - The @app.route() decorator maps a URL to a specific function.
          - When a user visits that URL, the connected function runs.
          - It is used to create endpoints for the API.
          - It helps define which HTTP methods the route accepts.
          - It is essential for building web routes in Flask.



- Question 5 : What is the role of Flask-SQLAlchemy?

          - Flask-SQLAlchemy allows Flask applications to work easily with databases.
          - It provides an ORM so we can use Python classes instead of SQL queries.
          - It helps create, read, update, and delete database records simply.
          - It supports relationships between tables.
          - It makes database management cleaner and more efficient.



- Question 6 : How do you create a basic Flask application?

          - First, import Flask and create an app object.
          - Then define a route using @app.route().
          - Inside the route, return a simple message or output.
          - Finally, run the app using app.run().
          - This creates a simple working Flask application.

          Example:
          from flask import Flask
          app = Flask(__name__)

          @app.route("/")
          def home():
              return "Hello Flask!"

          app.run()



- Question 7 : How do you return JSON responses in Flask?

          - Flask provides a function called jsonify() to return JSON data.
          - jsonify() converts Python dictionaries into JSON format automatically.
          - It also sets the correct content-type header.
          - JSON responses are commonly used in REST APIs.
          - Example: return jsonify({"message": "Success"})



- Question 8 : How do you handle POST requests in Flask?

          - To handle POST requests, specify methods=["POST"] in the route.
          - Use request.get_json() to read JSON data sent by the client.
          - The server processes the data and returns a response.
          - POST is used when sending data to the server.
          - Example:
            @app.route("/add", methods=["POST"])
            def add():
                data = request.get_json()
                return data



- Question 9 : How do you handle errors in Flask (e.g., 404)?

          - Flask allows custom error handling using @app.errorhandler().
          - You can return a custom JSON message for specific errors.
          - Example: 404 for “page not found”.
          - This makes the API more user-friendly and clear.
          - Example:
            @app.errorhandler(404)
            def not_found(e):
                return {"error": "Page not found"}, 404



- Question 10 : How do you structure a Flask app using Blueprints?

          - Blueprints help split a large application into smaller modules.
          - Each module has its own routes and logic.
          - Create a blueprint file and define routes inside it.
          - Register the blueprint in your main app using app.register_blueprint().
          - This makes the app organized, scalable, and easier to maintain.
